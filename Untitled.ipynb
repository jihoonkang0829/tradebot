```python
import gym_anytrading
from gym_anytrading.envs import StocksEnv, Actions, Positions
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
from assets.gru import GRU
```

```python
def my_calculate_reward(self, action):
    step_reward = 0

    current_price = self.prices[self._current_tick]
    if action == Actions.Buy.value:
        step_reward = current_price
    elif action == Actions.Sell.value:
        step_reward = -1 * current_price
    return step_reward


def mystep(self, action):
    self._done = False
    self._current_tick += 1
    if(action == 2):
        if self._position == Positions.Long:
            action = Actions.Buy.value
        elif self._position == Positions.Short:
            action = Actions.Sell.value
            
    if (self._current_tick) == self._end_tick:
        self._done = True

    step_reward = self._calculate_reward(action)
    self._total_reward += step_reward

    self._update_profit(action)

    if action == Actions.Buy.value:
        self._position = Positions.Long
    elif action == Actions.Sell.value:
        self._position = Positions.Short
    self._last_trade_tick = self._current_tick

    self._position_history.append(self._position)
    observation = self._get_observation()
    info = dict(
        total_reward = self._total_reward,
        total_profit = self._total_profit,
        position = self._position.value
    )
    self._update_history(info)

    return observation, step_reward, self._done, info

def my_update_profit(self, action):
    current_price = self.prices[self._current_tick]
    if action == Actions.Buy.value:
        self._total_profit = self._total_profit * (1+ (current_price * 10) )
    elif action == Actions.Sell.value:
        self._total_profit = self._total_profit * (1- (current_price * 10) )
        
def my_process_data(self):
    prices = self.df.loc[:, 'Close'].to_numpy()
    
    prices[self.frame_bound[0] - self.window_size]  # validate index (TODO: Improve validation)
    prices = prices[self.frame_bound[0]-self.window_size:self.frame_bound[1]]
    
    volumes = self.df.loc[:, 'volume'].to_numpy()
    
    volumes[self.frame_bound[0] - self.window_size]  # validate index (TODO: Improve validation)
    volumes = volumes[self.frame_bound[0]-self.window_size:self.frame_bound[1]]

    signal_features = np.column_stack((prices, volumes))

    return prices, signal_features

class MyFuturesEnv(StocksEnv):
    _calculate_reward = my_calculate_reward
    step = mystep
    _update_profit = my_update_profit
    _process_data = my_process_data
```

```python
model = torch.load('assets/gru_classification_model.pt')
```

```python
env = MyFuturesEnv(df = hourdf, window_size = 15, frame_bound = (int(0.8 * len(hourdf) / 10) * 10, len(hourdf) ))
observation = env.reset()
while True:
    observation = observation[np.newaxis, ...]
    print(observation.shape)
    observation = torch.from_numpy(observation).float()
    action = model(observation)
    observation, reward, done, info = env.step(action)

    if done:
        print("info:", info)
        break
```
